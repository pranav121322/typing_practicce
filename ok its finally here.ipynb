```python
import requests
import bs4
import random
import re
```

```python
def importing_words():
    link = "https://www.ef.com/wwen/english-resources/english-vocabulary/top-3000-words/"
    code = requests.get(link.format(1))
    #code.text
    final_code = bs4.BeautifulSoup(code.text,"html.parser")
    #final_code
    line = final_code.select('p')[-1]
    line = str(line)
    list1 = re.findall(r'[^\W]+',line)
    list1 = set(list1)
    list1 = list(list1)
    return list1
    #it give return as list
```

```python
def typing_string(list1):
    new_list = []
    for i in range(20):
        r1 = random.randint(0, 3000)
        new_list.append(list1[r1]) 
    str1 = "_".join(new_list)
    
    ans = " ".join(new_list)
    list_ans = []
    list_ans.append(str1)
    list_ans.append(ans)
    return list_ans
    # it gives string as return
    
```
